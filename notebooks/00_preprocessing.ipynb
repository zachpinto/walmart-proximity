```python
import pandas
from ydata_profiling import ProfileReport
```

```python
import pandas as pd

# Load the Walmart locations dataset
walmart_data_path = '/Users/pintoza/Desktop/dev/data-science/walmart-proximity/data/processed/walmart_locations.csv' 
walmart_df = pd.read_csv(walmart_data_path)

# Load the zip code details dataset
zipcode_data_path = '/Users/pintoza/Desktop/dev/data-science/walmart-proximity/data/processed/zip_code_classifications.csv'  # Replace with the actual file path
zipcode_df = pd.read_csv(zipcode_data_path)

# Display the first few rows of each dataset
print("Walmart Locations Data:")
print(walmart_df.head())

print("\nZip Code Details Data:")
print(zipcode_df.head())

# Basic information about the datasets
print("\nWalmart Locations Data Info:")
print(walmart_df.info())

print("\nZip Code Details Data Info:")
print(zipcode_df.info())

# Summary statistics for numerical fields
print("\nWalmart Locations Data Description:")
print(walmart_df.describe())

print("\nZip Code Details Data Description:")
print(zipcode_df.describe())

# Check for missing values
print("\nMissing Values in Walmart Locations Data:")
print(walmart_df.isnull().sum())

print("\nMissing Values in Zip Code Details Data:")
print(zipcode_df.isnull().sum())
```

```python
missing_ruca_df = zipcode_df[zipcode_df['ruca'].isnull()]
missing_ruca_df
```

```python
zipcode_df['ruca'].fillna('urban', inplace=True)
print(zipcode_df['ruca'].isnull().sum())
```

```python
zipcode_df['zip'] = zipcode_df['zip'].apply(lambda x: f"{int(x):05d}" if pd.notnull(x) else x)
zipcode_df.head()
```

```python
walmart_profile = ProfileReport(walmart_df, title="Walmart Profiling Report")
walmart_profile
```

```python
unique_designations = walmart_df['Designation'].unique()
unique_designations
```

```python
zip_code_profile = ProfileReport(zipcode_df, title="Zip Code Profiling Report")
zip_code_profile
```

```python
zipcode_df['ruca'] = zipcode_df['ruca'].replace('Suruban', 'Suburban')
zipcode_df['ruca'] = zipcode_df['ruca'].replace('urban', 'Urban')
```

```python
print("Updated unique values in 'ruca':")
print(zipcode_df['ruca'].unique())
```
